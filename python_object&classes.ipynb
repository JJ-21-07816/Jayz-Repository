```python
class calculator(): 
    
    def __init__(self, x, y):
        self.fnum = x
        self.snum = y

    def add(self):
        return self.fnum + self.snum


    def sub(self):
        return self.fnum - self.snum


    def mult(self):
        return self.fnum * self.snum


    def div(self):
        return self.fnum / self.snum


    def mod(self):
        return self.fnum % self.snum


    def exp(self):
        return self.fnum ** self.snum

while True:
    print("SELECT OPERATION FOR STANDARD AND SCIENTIFIC CALCULATOR:")
    print("")
    print("For 'ADDITION' Enter: 1")
    print("For 'SUBTRACTION' Enter: 2")
    print("For 'MULTIPLICATION' Enter: 3")
    print("For 'DIVISION' Enter: 4")
    print("For 'MODULUS' Enter: 5")
    print("For 'EXPONENT' Enter: 6")
    print("")
    
    choice = input("ENTER CHOICE (1/2/3/4/5/6): ")
    
    fnum = float(input("ENTER FIRST NUMBER: "))
    snum = float(input("ENTER SECOND NUMBER: "))
    operation = calculator(fnum, snum)
    
    print("")
    
    if choice in ('1', '2', '3', '4', '5', '6'):
              
        if choice == '1':
            print(fnum, " + ", snum, "=", operation.add())
            
        elif choice == '2':
            print(fnum, " - ", snum, "=", operation.sub())
                
        elif choice == '3':
            print(fnum, " * ", snum, "=", operation.mult())
                
        elif choice == '4':
            if snum == 0:
                print("Invalid! Cannot divide by Zero(0)!")
                continue
            else:
                print(fnum, " / ", snum, "=", operation.div())
                    
        elif choice == '5':
            print(fnum, " % ", snum, "=", operation.mod())
                
        elif choice == '6':
            print(fnum, " ^ ", snum, "=", operation.exp())
        
        print("")
        print("---'Y' for Yes & 'N' for No---")
        again = input("Would you like to Try Again?: ")
        if again in ('Y', 'N'):
            if again == 'Y':
                continue
            if again == 'N':
                print("Thank You! Have a nice day!")
                break
    else:
        print("INVALID INPUT")
        
```
