```python
print("========= PYTHON VARIABLES =========")

print()
print("=========================================================================================================================================================")
print()

print("A variable is a named location used to store data in the memory.")

print("# Creating Variables.")
print()
print("Integer variables can be declared by assigning a value to it in a single declaration.")

number = 10

print("This is a number", number,)
print()
print("String variables can be declared either by using single or double quotes")

name = "John"

print("This is a string", name)
print()
print("We can think that the variable is a kind of container to store items in it and that items can be replaced at anytime.")

number = 15

print("From ' number = 10 ', To ' number =", number, "'.")

name = "Joshua"

print("From ' name = John ', To ' name =", name, "'.")

print()
print("=========================================================================================================================================================")
print()

print("# Variables are Case-Sensitive.")
print()

a = 20
A = "Twenty"

print("\"A\" variable will not overwrite variable \"a\"")
print("Variable a =", a)
print("Variable A =", A)

print()
print("=========================================================================================================================================================")
print()

print("# Variables Naming.")
print()
print("1. Variable names should have a combination of letters in lowercase (a to z) or uppercase (A to Z) or digits (0 to 9) or an underscore (_).")
print("For Example:")

alpha_numeric = "alpha_numeric"
ALPHA_NUMERIC = "ALPHA_NUMERIC"
alphaNumeric = "alphaNumeric"
AlphaNumeric = "AlphaNumeric"

print("snake_case =", alpha_numeric, "\nMACRO_CASE =", ALPHA_NUMERIC, "\ncamelCase =", alphaNumeric, "\nCapWords =", AlphaNumeric)
print()
print("2. Create a name that makes sense.")

alpha = "same"
a = "same"

print("For Example:\nVariable named alpha =", alpha, ",make sense than Variable named a =", a)
print()
print("3. Never use special symbols like !, @, #, $, %, etc.")

print("For Example:\n!name, @name, #name, $name, %name, etc")

print()
print("4. Don't start a variable name with a digit.")

print("For Example:\n123name, 0name, 321name")

print()
print("=========================================================================================================================================================")
print()

print("# Assigning multiple values.")
print()

a, b, c = "Tom", "and", "Jerry"
print("For example:") 
print("a =", a, "| b =", b, "| c =", c)

print()

a = b = c = "Tom and Jerry"
print("For example:")
print("a =", a)
print("b =", b)
print("c =", c)
```

```python
print("========= PYTHON DATA TYPES =========")

print()
print("=========================================================================================================================================================")
print()

print("# Data Type Examples.")
print()

this_str = "John Joshua"
this_int = 20
this_float = 20.5
this_complex = 5j
this_list = ["Tanjiro", "Inosuke", "Zenitsu"]
this_tuple = ("One", "Two", "Three")
this_range = range(20)
this_dict = {"Name" : "John", "Age" : "20"}
this_set = {"Jupiter", "Saturn", "Uranus"}
this_bool = True

print("String =", this_str)
print("Integer =", this_int)
print("Float =", this_float)
print("Complex =", this_complex)
print("List =", this_list)
print("Tuple =", this_tuple)
print("Range =", this_range)
print("Dictionary =", this_dict)
print("Set =", this_set)
print("Boolean =", this_bool)

print()
print("=========================================================================================================================================================")
print()

print("# Specifying Data Types.")
print()

sample_string = str("John Joshua")
sample_int = int(20)

print("Hi my name is", sample_string, "and I am", sample_int, "years old.")

print()
print("=========================================================================================================================================================")
print()

print("# Getting its Data Types.")
print()

name = "John Joshua Bartolome"
number = 20

print(type(name))
print(type(number))
```

```python
print("========= PYTHON OPERATORS =========")

print()
print("=========================================================================================================================================================")
print()

print("# Arithmetic Operators.")
print()

x = 10
y = 5

print("Addition =", x, "+", y, "=", x + y)
print("Subtraction =", x, "-", y, "=", x - y)
print("Multiplication =", x, "*", y, "=", x * y)
print("Division =", x, "/", y, "=", x / y)
print("Modulus =", x, "%", y, "=", x % y)
print("Exponentiation =", x, "**", y, "=", x ** y)
print("Floor Division =", x, "//", y, "=", x // y)


```
